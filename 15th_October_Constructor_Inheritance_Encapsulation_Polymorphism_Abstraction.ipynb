<b>Runtime Polymorphism:</b> Runtime polymorphism, also known as dynamic polymorphism or late binding, refers to the resolution of function calls at runtime. The appropriate method to be called is determined by the actual type of the object at runtime<br>
In Python, runtime polymorphism is commonly achieved through inheritance and method overriding. When a subclass overrides a method from its superclass, the method call on an object of the subclass will invoke the overridden method in the subclass, rather than the method in the superclass. This allows objects of different classes to be treated uniformly when calling the same method<br>

It promotes code flexibility and reusability by allowing the reuse of existing classes as components within new classes. It also reduces coupling between objects, making the codebase less rigid and more adaptable to change.<br>
In summary, while composition has some challenges and drawbacks, it remains a valuable tool for designing software systems. The concept of "aggregation" in composition refers to a relationship between two classes that denotes a "has-a" relationship, where one class is composed of one or more instances of another class. Aggregation is a form of association that represents a whole-part relationship, where the part can exist independently of the whole. It is a weaker form of association compared to composition.<br>
Here are the key differences between aggregation and composition:
Independence: In aggregation, the child object can exist independently of the parent object. However, in composition, the child object is dependent on the parent object and cannot exist independently.

Lifetime: In aggregation, the lifetime of the associated objects is independent of each other. In composition, the lifetime of the associated objects is tied together, and if the parent object is destroyed, the child objects are also destroyed.

Representation: In UML diagrams, aggregation is represented by a straight line with an empty diamond at one end, while composition is represented by a straight line with a filled diamond at one end.

Dependency: In aggregation, the associated objects are not highly dependent on each other, whereas in composition, the objects are highly dependent on each other.<br>

In summary, aggregation and composition are both types of association used to represent relationships between classes. Aggregation represents a weaker association where the associated objects can exist independently, while composition represents a stronger association where the associated objects are highly dependent on each other.<br>

This allows for easy replacement or modification of dependencies at runtime. Instead of instantiating objects directly, you can pass them as arguments to the constructor or setter methods. This way, you can easily replace or modify the dependencies by providing different objects when creating or updating the composed object.

Interfaces: By defining interfaces for the composed objects, you can create interchangeable components that share common behavior. This allows you to replace or modify the objects dynamically at runtime by providing different implementations of the same interface. The composed object can use the interface methods without knowing the specific implementation details.

Polymorphism: Polymorphism allows objects to be treated as instances of their parent classes or interfaces. By using polymorphism, you can replace or modify composed objects dynamically with other objects that inherit from the same parent class or implement the same interface. This allows you to switch between different implementations of the composed objects at runtime without affecting the functionality of the overall system.<br>
